```python
st_vozlisc = 1000
st_povezav =  1000

graf = graphs.RandomGNM(st_vozlisc, st_povezav)

#ILP
ILP = MixedIntegerLinearProgram(maximization = False)
vozlisca = ILP.new_variable(binary = True)
ILP.set_objective( sum([vozlisca[v] for v in graf]) )

for u,v in graf.edges(labels = False):
    ILP.add_constraint( vozlisca[u] + vozlisca[v] >= 1 )

ILP.solve()
vozlisca = ILP.get_values(vozlisca)
print [v for v,i in vozlisca.items() if i]

#Relaxacija na LP
LP = MixedIntegerLinearProgram(maximization = False)
vozlisca_LP = LP.new_variable(real = True)
LP.set_max(vozlisca_LP,1)
LP.set_min(vozlisca_LP,0)
LP.set_objective( sum([vozlisca_LP[v] for v in graf.vertices()]) )

for u,v in graf.edges(labels = False):
    LP.add_constraint( vozlisca_LP[u] + vozlisca_LP[v] >= 1 )


LP.solve()
vozlisca_LP = LP.get_values(vozlisca_LP)
print [v for v,i in vozlisca_LP.items() if i>=1/2]

#Greedy
graf.vertex_cover(algorithm='greedy', value_only=True)
```
